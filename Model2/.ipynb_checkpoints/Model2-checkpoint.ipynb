```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
import seaborn as sbn
import numpy as np
```

```python
data_frame=pd.DataFrame(pd.read_csv("zomato.csv"))
```

```python
data_frame.isnull().sum()
data_frame
```

```python
sbn.displot(data_frame["Volume"])
```

```python
data_frame.corr()["Volume"].sort_values
```

```python
data_frame=data_frame.sort_values("Volume",ascending=False).iloc[13:]
```

```python
data_frame
```

```python
data_frame.drop("Date",axis=1,inplace=True)
y=data_frame["Volume"].values
x=data_frame.drop("Volume",axis=1).values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=15)
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
model=Sequential()
model.add(Dense(6))
model.add(Dropout(0.5))
model.add(Dense(6))
model.add(Dropout(0.5))
model.add(Dense(1))
model.compile(optimizer="adam",loss="mse")
early_stopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1)
model.fit(x=x_train,y=y_train,epochs=100)
```

```python
loss=pd.DataFrame(model.history.history)
```
